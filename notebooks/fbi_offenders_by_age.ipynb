```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

input_path = '../data/cleandata/fbi_offenders_by_age.pkl'

fbi_data = pd.read_pickle(input_path)
fbi_data
```

```python
#Make Ages One Column
fbi_data_year = fbi_data['data_year'].values.tolist()
fbi_data_offense = fbi_data['offense_type_list'].values.tolist()
age_ranges = ['unknown','range_0_9','range_10_19','range_20_29','range_30_39','range_40_49','range_50_59','range_60_69','range_70_79','range_80_89','range_90_99']

for age_range in age_ranges:
    fbi_data_offender_count = fbi_data[age_range].values.tolist()
    fbi_data_offender_age_range = [None]*fbi_data.shape[0]
    
    offender_index = 0
    for record in fbi_data_offender_age_range:
        fbi_data_offender_age_range[offender_index] = age_range
        offender_index += 1
    
    fbi_data_temp = pd.DataFrame({'Year':fbi_data_year, 'Offense':fbi_data_offense, 'Offenders':fbi_data_offender_count, 'Age Range':fbi_data_offender_age_range})
    if age_range == 'unknown':
        fbi_data_compressed = fbi_data_temp
    else:
        fbi_data_compressed = pd.concat([fbi_data_compressed,fbi_data_temp])        

fbi_data_compressed.reset_index(drop=True)
```

```python
#Group By Offense Type
fbi_data_by_age_by_offense = fbi_data_compressed.groupby(['Offense','Age Range'])
fbi_data_by_age_by_offense = pd.DataFrame(fbi_data_by_age_by_offense['Offenders'].sum())
fbi_data_by_age_by_offense = fbi_data_by_age_by_offense.reset_index()
fbi_data_by_age_by_offense

offense_pivot = pd.pivot_table(fbi_data_by_age_by_offense,values='Offenders',index=['Offense'],columns=['Age Range'],aggfunc=np.sum)
offense_pivot = offense_pivot.reset_index()
offense_pivot
```

```python
offense_pivot['Total']=offense_pivot['range_0_9']+offense_pivot['range_10_19']+offense_pivot['range_20_29']
offense_pivot['Total']=offense_pivot['Total']+offense_pivot['range_30_39']+offense_pivot['range_40_49']
offense_pivot['Total']=offense_pivot['Total']+offense_pivot['range_50_59']+offense_pivot['range_60_69']
offense_pivot['Total']=offense_pivot['Total']+offense_pivot['range_70_79']+offense_pivot['range_80_89']
offense_pivot['Total']=offense_pivot['Total']+offense_pivot['range_90_99']
offense_pivot['range_0_9%']=offense_pivot['range_0_9']/offense_pivot['Total']*100
offense_pivot['range_10_19%']=offense_pivot['range_10_19']/offense_pivot['Total']*100
offense_pivot['range_20_29%']=offense_pivot['range_20_29']/offense_pivot['Total']*100
offense_pivot['range_30_39%']=offense_pivot['range_30_39']/offense_pivot['Total']*100
offense_pivot['range_40_49%']=offense_pivot['range_40_49']/offense_pivot['Total']*100
offense_pivot['range_50_59%']=offense_pivot['range_50_59']/offense_pivot['Total']*100
offense_pivot['range_60_69%']=offense_pivot['range_60_69']/offense_pivot['Total']*100
offense_pivot['range_70_79%']=offense_pivot['range_70_79']/offense_pivot['Total']*100
offense_pivot['range_80_89%']=offense_pivot['range_80_89']/offense_pivot['Total']*100
offense_pivot['range_90_99%']=offense_pivot['range_90_99']/offense_pivot['Total']*100
offense_pivot['Under 30 Years %'] = offense_pivot['range_0_9%']+offense_pivot['range_10_19%']+offense_pivot['range_20_29%']
offense_pivot['Over 30 Years %'] = offense_pivot['range_30_39%']+offense_pivot['range_40_49%']+offense_pivot['range_50_59%']
offense_pivot['Over 30 Years %'] = offense_pivot['Over 30 Years %']+offense_pivot['range_60_69%']+offense_pivot['range_70_79%']
offense_pivot['Over 30 Years %'] = offense_pivot['Over 30 Years %']+offense_pivot['range_80_89%']+offense_pivot['range_90_99%']
offense_pivot
```

```python
offender_total_by_age = fbi_data_compressed.groupby(['Age Range'])
offender_total_by_age = pd.DataFrame(offender_total_by_age['Offenders'].sum())
offender_total_by_age = offender_total_by_age.reset_index()
offender_total_by_age
```

```python
labels = offender_total_by_age['Age Range']
values = offender_total_by_age['Offenders']

plt.pie(values,labels=labels,autopct="%1.1f%%",shadow=True,startangle=180,pctdistance=0.6)
plt.show()
```

```python
offense_pivot_young_crimes = offense_pivot.sort_values(by='Under 30 Years %', ascending=False)
top_young_crimes = offense_pivot_young_crimes['Offense']
top_young_crimes = top_young_crimes[0:20]
crime_range = range(0,20)
offense_pivot_young_crimes = offense_pivot[offense_pivot['Offense'].isin(top_young_crimes)]
offense_pivot_young_crimes = offense_pivot_young_crimes.sort_values(by='Under 30 Years %', ascending=False)
offense_pivot_young_crimes
```

```python
plt.bar(crime_range, offense_pivot_young_crimes['Under 30 Years %'].values.tolist(), width=0.8, label='Under 30', color='salmon')
plt.bar(crime_range, offense_pivot_young_crimes['Over 30 Years %'].values.tolist(), width=0.8, label='Over 30', color='skyblue', bottom = offense_pivot_young_crimes['Under 30 Years %'].values.tolist())

plt.xticks(crime_range, top_young_crimes)
plt.rcParams["figure.figsize"] = [12,6]
plt.xticks(rotation=45)
plt.xticks(horizontalalignment="right")
plt.legend(loc='best')

plt.show()
```
